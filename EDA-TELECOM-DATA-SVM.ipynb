```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
%matplotlib inline
```

```python
df =  pd.read_csv('./Churn_Modelling.csv')
df.head()
```

```python
len(df)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df["Age"].mean())
df.isnull().sum()
```

```python
df.head()
```

```python
(df["Exited"] == 1).sum()
```

```python
(df["Exited"] == 0).sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df["Geo_class"] = label_encoder.fit_transform(df["Geography"])
df["Gender_class"] = label_encoder.fit_transform(df["Gender"])
```

```python
df.head(101)
```

```python
X = df.drop(["Exited", "Geography", "Gender", "Surname", "Geo_class", "Gender_class"], axis = ('columns'))
```

```python
Y = df.Exited
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.05 ,random_state = 44)
```

```python
model = svm.SVC(kernel = 'rbf')
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
